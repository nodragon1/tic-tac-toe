```python
class tictactoe:
    
    def __init__(self):
        self.board = []
    
    def create_board(self):
        for i in range(3):
            row = []
            for j in range(3):
                row.append('-')
            self.board.append(row)
    
    def show_board(self):
        for row in self.board:
            for item in row:
                print(item, end=' ')
            print()
    
    def is_player1_win(self):
        for i in range(3):
            count = 0
            for j in range(3):
                if self.board[i][j] == 'O':
                    count += 1
            if count == 3:
                return True

        for i in range(3):
            count = 0
            for j in range(3):
                if self.board[j][i] == 'O':
                    count += 1
            if count == 3:
                return True

        count = 0
        for i in range(3):
            if self.board[i][i] == 'O':
                count += 1
        if count == 3:
            return True

        count = 0
        for i in range(3):
            if self.board[2 - i][i] == 'O':
                count += 1
        if count == 3:
            return True
        return False
    
    def is_player2_win(self):
        for i in range(3):
            count = 0
            for j in range(3):
                if self.board[i][j] == 'X':
                    count += 1
            if count == 3:
                return True

        for i in range(3):
            count = 0
            for j in range(3):
                if self.board[j][i] == 'X':
                    count += 1
            if count == 3:
                return True

        count = 0
        for i in range(3):
            if self.board[i][i] == 'X':
                count += 1
        if count == 3:
            return True

        count = 0
        for i in range(3):
            if self.board[2 - i][i] == 'X':
                count += 1
        if count == 3:
            return True
        return False
    
    def is_board_filled(self):
        for row in self.board:
            for item in row:
                if item == '-':
                    return False
        return True
    
    def start_game(self):
        self.create_board()
        n = 0
        while True:
            self.show_board()
            if n % 2 == 0:
                print('Player1 Turn')
                x, y = map(int, input())
                if self.board[x][y] == '-':
                    self.board[x][y] = 'O'
                    if self.is_player1_win():
                        self.show_board()
                        print('Player1 Wins!')
                        break
                    if self.is_board_filled():
                        if self.is_player1_win():
                            self.show_board()
                            print('Player1 Wins!')
                        else:
                            self.show_board()
                            print('Draw. Game Over.')
                        break
                    n += 1
                else:
                    print('Try Again!')
            else:
                print('Player2 Turn')
                x, y = map(int, input())
                if self.board[x][y] == '-':
                    self.board[x][y] = 'X'
                    if self.is_player2_win():
                        self.show_board()
                        print('Player2 Wins!')
                        break
                    if self.is_board_filled():
                        if self.is_player2_win():
                            self.show_board()
                            print('Player2 Wins!')
                        else:
                            self.show_board()
                            print('Draw. Game Over.')
                        break
                    n += 1
                else:
                    print('Try Again!')
```

```python
game = tictactoe()
game.start_game()
```
